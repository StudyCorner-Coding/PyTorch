# **Deep Learning**

Deep learning is a subset of machine learning that uses neural networks with multiple layers to model complex patterns in data. PyTorch is one of the most popular deep learning frameworks, offering flexibility, ease of use, and dynamic computation graphs.

# **Foundations of Deep Learning**

# **1. Neural Networks Basics**


**Perceptron**: The simplest neural network unit, inspired by biological neurons.



**Activation Functions**: Functions like ReLU, Sigmoid, and Tanh introduce non-linearity.



**Loss Functions**: Measures error (e.g., MSE for regression, Cross-Entropy for classification).



**Optimization**: Algorithms like Stochastic Gradient Descent (SGD), Adam, and RMSprop update model weights.

# **2. Deep Neural Networks**


**Feedforward Networks:** Simple multi-layer networks.



**Backpropagation:** Algorithm to update weights using gradients.



**Regularization:** Techniques like dropout and L2 regularization prevent overfitting.

# **3. Convolutional Neural Networks (CNNs)**


CNN Used for image processing.



**Convolution Layers:** Extract spatial features.



**Pooling Layers:** Reduce dimensionality.



**Fully Connected Layers**: Flatten and classify.

# **4. Recurrent Neural Networks (RNNs) & Transformers**


RNNs handle **sequential** data.



**LSTMs & GRUs:** Solve vanishing gradient problem in RNNs.



**Transformers:** Foundation for models like BERT and GPT

# **PyTorch: A Deep Learning Framework**

# **What is PyTorch ?**

PyTorch is an open source machine learning library for Python and is completely based on Torch.

It is primarily used for applications such as natural language processing.

PyTorch is developed by Facebook's artificial-intelligence research group along with Uber's "Pyro" software for the concept of in-built probabilistic programming.

PyTorch is an open-source deep learning framework that's known for its flexibility and ease-of-use.


This is enabled in part by its compatibility with the popular python high-level programming language favoured by machine learning developers and data scientists

# **History of PyTorch**

 PyTorch was developed by Hugh Perkins as a Python wrapper for the LusJIT based on Torch framework.There are two PyTorch variants.



PyTorch redesigns and implements Torch in Python while sharing the same core C libraries for the backend code.

PyTorch developers tuned this back-end code to run Python efficiently. They also kept the GPU based hardware acceleration as well as the extensibility features that made Lua-based Torch.

# **Why PyTorch ?**

1. **Dynamic** **computation** **graphs**.

2. **Easy** **debugging** and **intuitive** **API**.

3. **Strong GPU support with CUDA.**

# **Features**


The major features of PyTorch are mentioned below −



**Easy Interface** − PyTorch offers easy to use API; hence it is considered to be very simple to operate and runs on Python. The code execution in this framework is quite easy.



**Python usage** − This library is considered to be Pythonic which smoothly integrates with the Python data science stack. Thus, it can leverage all the services and functionalities offered by the Python environment.



**Computational graphs** − PyTorch provides an excellent platform which offers dynamic computational graphs. Thus a user can change them during runtime. This is highly useful when a developer has no idea of how much memory is required for creating a neural network model.

# **Core Concepts**


1. **Tensors:** Multi-dimensional arrays (like NumPy but optimized for GPUs).

2. **Autograd:** Automatic differentiation for backpropagation.

3. **nn.Module:** Base class for all neural networks.

4. **DataLoader:** Efficient data loading and batching.

5. **CUDA:** GPU acceleration with torch.cuda.
